```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = (
    SparkSession.builder
    .appName("Exploracao_SQL")
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .config(
        "spark.sql.catalog.spark_catalog",
        "org.apache.spark.sql.delta.catalog.DeltaCatalog",
    )
)

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
df_gold = spark.read.format("delta").load("src/de_etl_pipeline_loja/data/gold/vendas_produto_mensal")
df_gold.createOrReplaceTempView("vendas_produto_mensal")

spark.sql("""
    SELECT ano, mes, produto, receita_total
    FROM vendas_produto_mensal
    WHERE ano = 2024
    ORDER BY mes, receita_total DESC
    LIMIT 10
""").show()
```
